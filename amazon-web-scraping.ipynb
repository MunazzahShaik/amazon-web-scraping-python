```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#function to extract product title

def get_title(soup):
    try:
        Title = soup.find("span", attrs = {"id":'productTitle'}).text.strip()
    
    except AttributeError :
        Title = " "
    
    return Title

#function to extract product title

def get_price(soup):
    try:
        Price = soup.find("span", attrs = {"class" : 'a-offscreen'}).text
    
    except:
        Price = " "
        
    return Price
        
#function to extract rating 

def get_rating(soup):
    try:
        Rating = soup.find("span", attrs={"class":'a-icon-alt'}).text.strip().split()[0]
    
    except :
        Rating = " "
    
    return Rating



# Function to extract Availability Status
def get_deliverydate(soup):
    try:
        delivery_date = new_soup.find("div", attrs={'id':'mir-layout-DELIVERY_BLOCK'}).find("span", attrs = {"class": "a-text-bold"}).text.strip()
    
    except AttributeError :
        delivery_date = " "
    
    return delivery_date


def get_availability(soup):
    try:
        availability = new_soup.find("div", attrs={'id':'availability'}).find("span").text.strip()

    except AttributeError:
        availability = "Not Available"

    return availability
```

```python
links_list = []

    # Loop for extracting links from Tag Objects
for link in links:
        links_list.append(link.get('href'))

d = {"Title":[], "Price":[], "Rating":[], "Availability":[], "Delivery Date":[]}
    
    # Loop for extracting product details from each link 
for link in links_list:
        new_webpage = requests.get("https://www.amazon.in" + link, headers=HEADERS)

        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        # Function calls to display all necessary product information
        d['Title'].append(get_title(new_soup))
        d['Price'].append(get_price(new_soup))
        d['Rating'].append(get_rating(new_soup))
        d['Availability'].append(get_availability(soup))
        d['Delivery Date'].append(get_deliverydate(soup))
        
```

```python
amazon_df = pd.DataFrame.from_dict(d)
amazon_df['Title'].replace('', np.nan, inplace=True)
amazon_df = amazon_df.dropna(subset=['Title'])
amazon_df.to_csv("amazon_data.csv", header=True, index=False)
```
